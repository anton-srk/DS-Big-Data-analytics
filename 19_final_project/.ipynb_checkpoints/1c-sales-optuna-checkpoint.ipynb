```python
import numpy as np
import pandas as pd
import dateutil
import datetime as dt
from datetime import date
import matplotlib.pyplot as plt
import seaborn as sns
import re
import gc
from tqdm.notebook import tqdm
import pickle

from itertools import permutations, product

import lightgbm as lgb

from lightgbm import LGBMRegressor
from lofo import LOFOImportance, FLOFOImportance, Dataset, plot_importance

from sklearn.metrics import (mean_squared_log_error, mean_squared_error,
                             mean_absolute_error)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

from sklearn.model_selection import train_test_split, KFold
import warnings

import optuna
```

## Data preparation

```python
alles = pd.read_pickle('data/alles_final.pkl')
```

```python
rows = alles['date_block_num'] > 11
```

```python
features = alles.columns[(alles.columns!='item_cnt_month')]
```

```python
data = alles.loc[rows, list(features)+['item_cnt_month']]
data = data.fillna(0)
```

Trying different train/test combinations here:

```python
train_idx, test_idx = train_test_split(list(range(12, 34)), test_size=0.2)

train_rows = data['date_block_num'].isin(train_idx)
test_rows = data['date_block_num'].isin(test_idx)

# train_rows = (data.date_block_num <= 33)&(data.date_block_num >= 26)
# test_rows = data.date_block_num < 26

# train_rows = data.date_block_num < 33
# test_rows = data.date_block_num == 33
```

```python
X_train = data[train_rows].drop(['item_cnt_month'], axis=1)
y_train = data[train_rows]['item_cnt_month']
X_test = data[test_rows].drop(['item_cnt_month'], axis=1)
y_test = data[test_rows]['item_cnt_month']

X_final_test = data[data.date_block_num == 34].drop(['item_cnt_month'], axis=1)
```

```python
del alles, data

gc.collect()
```

Two variants of the parameter batches:

```python
# FYI: Objective functions can take additional arguments
# (https://optuna.readthedocs.io/en/stable/faq.html#objective-func-additional-args).
def objective(trial):

    d_train = lgb.Dataset(X_train, y_train)
    
    param = {
    'objective': 'regression',
    'metric': 'mse',
    'verbosity': -1,
    'boosting_type': 'gbdt',
    'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),
    'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0),
    'num_leaves': trial.suggest_int('num_leaves', 40, 256),
    'feature_fraction': trial.suggest_uniform('feature_fraction', 0.4, 1.0),
    'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.4, 1.0),
    'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
    'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
    }
    
#     param = {
#     'objective': 'regression',
#     'verbosity': -1,
#     'metric': 'rmse',
#     'boosting_type': 'gbdt',
#     'max_depth': trial.suggest_int('max_depth', 2, 12),
#     'num_leaves': trial.suggest_int('num_leaves', 40, 256),
#     'min_child_weight': trial.suggest_uniform('min_child_weight', 0.25, 0.75), 
#     'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.6, 1.0), 
#     'subsample': trial.suggest_uniform('subsample', 0.6, 1.0)
#     }

    gbm = lgb.train(param, d_train)
    preds = gbm.predict(X_test)
    mse = mean_squared_error(y_test, preds)
    return mse
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=100)

print('Number of finished trials: {}'.format(len(study.trials)))

print('Best trial:')
trial = study.best_trial

print('  Value: {}'.format(trial.value))

print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))
```

```python
with open('data/final_params.pkl', 'wb') as handle:
    pickle.dump(trial.params, handle)
```
